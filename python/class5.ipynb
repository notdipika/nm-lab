```python
import csv

data = [
    ['Name', 'Address', 'Age'],
    ['Ram', 'Itahari', '23'],
    ['Shyam', 'Dharan', '22'],
    ['Hari', 'Biratnagar', '21']
]

with open(file="newnew.csv", mode='w', newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
import csv

data = [
    {'Name': 'Ram', 'Address': 'Itahari', 'Age': '23'},
    {'Name': 'Shyam', 'Address': 'Dharan', 'Age': '22'},
    {'Name': 'Hari', 'Address': 'Biratnagar', 'Age': '21'}
]

with open(file='newnewnew.csv', mode='w', newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["Name", "Address", "Age"])
    writer.writeheader()
    writer.writerows(data)
```

```python
import pandas as pd


df = pd.read_csv("newnewnew.csv")
print(df)
print(type(df))
```
