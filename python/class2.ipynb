```python
#factorial using while loop

f = 1
n = 5
while(True):
    if n <= 1:
        break
    f = f * n
    n = n - 1
print(f)

```

```python
#sum of n natural numbers
def sum(n):
    total = 0
    for i in range(1, n+1):
        total += i
    return total
print(sum(1000))
```

```python
#write a function named average which computes average of n numbers passed as arguments
def average(*args):
    s = 0
    for num in args:
        s = s + num
    return s/len(args)

average(1,2,3,4,5)
```

```python
def sum(**kwargs):
    total = 0
    for key in kwargs.keys():
        total += kwargs[key]
    return total

sum(a=10, b=20)
```

```python
class Point:
    def __init__(self, xx, yy):
        self.x = xx
        self.y = yy

    def __str__(self):
        return f'({self.x}, {self.y})'
    
    def __add__(self, other):
        new_x = self.x +other.x
        new_y = self.y +other.y
        return Point(new_x,new_y)


    
p1 = Point(10,20)
p2 = Point(30,40)
p3 = p1 +p2 #p1.__add__(p2)
print(p3)
```
