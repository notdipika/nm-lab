```python
# Base class
class Animal:
    def __init__(self,name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} is speaking")

# Derived Class
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name) #super calls the base class 
        self.breed = breed
    
    def speak(self):
        super().speak()  # Polymorphism
        print(f"{self.name} of {self.breed} is Barking!")

d1= Dog("Lucy", "Shiba Inu")
d1.speak()
    
```

```python
# Multiple inheritance
class A:
    def show(self):
        print("A")

class B:
    def show(self):
        print("B")

class C(A,B):
    pass

c1 = C()
c1.show()
```

```python
# Method Resolution order
C.mro()
```

```python
class DataList:
    def __init__(self, alist):
        self.alist = alist

    def mean(self):
        return sum(self.alist)/len(self.alist)
```

```python
l1 = DataList([1,2,3,4,5,6,7,8])
l1.mean()
```

```python
#Exception Handling
while(True):
    try:
        n1 = input("First Number? ")
        n2 = input("Second Number? ")
        if n1 == 'q' or n2 == 'q'
            break
        result = int(n1/n2)
        print("Result is ", result)
    except ZeroDivisionError:
        print("Cannot divide by zero")
    except ValueError:
        print("Please enter valid number")
    finally:
        print("This always runs")
```
